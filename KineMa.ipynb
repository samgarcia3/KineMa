### A User Independent System

We'll start with building recommendations for the case where we have no knowledge about a user. In this case, we probably would find the most popular items, and recommend them to our user.

The simplest system would just sort the items based on their average score and present a list in that order:

```python
# Sort movies by rating
top_rated = movie_df.sort_values(by=['vote_average'], ascending=False)
top_rated['title'].head(10)
```

```python
top_rated.head(10)
```

I can notice one large issue with the average score within this system. The majority of perfect average scores have very low vote counts and should not be included when being compared with films that have 100+ reviews(`vote_count`)

Take a look at a popular film and understand how many votes that film has and what the average rating is for that film to attain a better understanding 'realistically' of how many reviews a film should have and what the average rating would be.

<br>

The movie that quickly comes to mind for me is ["Ex Machina"](https://www.imdb.com/title/tt0470752/)

- let's bring up the information for that particular film.

```python
# Retrieve film Ex machina
top_rated[top_rated['title']== "Ex Machina"]
```

Ex Machina is a widely popular movie and as we can see has a total vote_count of 4,862 and this number is more indicative of the number of reviews a film should be recieving.
<br>

I will deal with the films with minimal reviews by creating a threshold in which we ignore items with too few reviews.

```python
# Initiate threshold count
threshold = 1000
# Retrieve movies that break the threshold
movies_with_many_reviews_df = movie_df[movie_df['vote_count'] >= threshold]

# Create new dataframe for movies above 1k votes
top_rated_v2 = movies_with_many_reviews_df.sort_values(by=['vote_average'], ascending=False)
top_rated_v2.head(10)
```
