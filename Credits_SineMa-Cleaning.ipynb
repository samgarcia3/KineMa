## Metadata Cleaning - (Crew)

```python
credits = pd.read_csv('data/credits.csv')
keywords = pd.read_csv('data/keywords.csv')
metadata = pd.read_csv('data/movies_metadata.csv')
```

```python
metadata['genres'] = metadata['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
metadata['year'] = pd.to_datetime(metadata['release_date'], errors='coerce').apply(lambda x: str(x).split('-')[0] if x != np.nan else np.nan)
```

```python
metadata = metadata.drop([19730, 29503, 35587])
```

```python
keywords['id'] = keywords['id'].astype('int')
credits['id'] = credits['id'].astype('int')
metadata['id'] = metadata['id'].astype('int')
```

```python
metadata = metadata.merge(credits, on='id')
metadata = metadata.merge(keywords, on='id')
```

**CREW**: Most recommendation systems that I have come across disregard the crew and only keep the Director for their system - thinking that the director is the sole proprieter of the "Feel" of a film. This is wrong, and I cannot stress enough that a film director highly relies on the artisanship and expertise of many of those that make up a film crew. Such as:

<br>

- Cinematographer
- Production Designer
- Lighting
- Hair / Make Up / Wardrobe
- Editor
- Colorist

and that is to name just a few. Film directors do not make movies alone, filmmaking is a highly collaborative effort and each department leans on one another. Directors, the good directors are those that can be humble and take the advice from the departments heads on and off set.

<br>

Take for instance - "Jaws" - Directed by Steven Spielberg. Was it not for the great Verna Fields who had been a long time editor working with the likes of Alfred Hitchcock, there would be no "Jaws" today. She understood and learned from the great suspense director Alfred Hitchcock that if you only enveil the "Horror" for a split second and let the viewer's mind do the rest - that it can actually be even more terrorfying than showing the "Horror" completely. This is was Verna taught Steven the day he returned to Universal Studios after a disastrous shoot day in which they could not get the mechainal shark to work. The unsung heroes like Verna are an incredibly important part to filmmaking as they bring the combined magic of those that came before them.
